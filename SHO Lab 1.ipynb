```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as ax
```

```python
h = .0001
n = 100000
t_max = h * n
print(t_max)
```

```python
x1 = np.zeros(n)
v1 = np.zeros(n)
t1 = np.zeros(n)
```

```python
x1[0] = 1
v1[0] = 0
t1[0] = 0
```

```python
for i in range(n-1):
    dx1 = v1[i] * h
    dv1 = -9*x1[i] * h
    
    x1[i+1] = x1[i] + dx1
    v1[i+1] = v1[i] + dv1
    t1[i+1] = t1[i] + h
```

```python
plt.figure()
plt.grid()
plt.plot(t1,x1)
plt.show()
```

```python
x2 = np.zeros(n)
v2 = np.zeros(n)
t2 = np.zeros(n)
```

```python
x2[0] = 1
v2[0] = 0
t2[0] = 0
```

```python
for i in range(n-1):
    dx2 = v2[i] * h
    dv2 = (-2 * v2[i] - 16 * x2[i]) * h
    
    x2[i+1] = x2[i] + dx2
    v2[i+1] = v2[i] + dv2
    t2[i+1] = t2[i] + h
```

```python
plt.figure()
plt.grid()
plt.plot(t2,x2)
plt.show()
```

```python
x3 = np.zeros(n)
v3 = np.zeros(n)
t3 = np.zeros(n)
```

```python
x3[0] = 1
v3[0] = 0
t3[0] = 0
```

```python
for i in range(n-1):
    dx3 = v3[i] * h
    dv3 = (-12 * v3[i] - 3 * x3[i]) * h
    
    x3[i+1] = x3[i] + dx3
    v3[i+1] = v3[i] + dv3
    t3[i+1] = t3[i] + h
```

```python
plt.figure()
plt.grid()
plt.plot(t3,x3)
plt.show()
```

```python
fig, ax = plt.subplots()
plt.plot(t1,x1)
plt.plot(t2,x2)
plt.plot(t3,x3)
ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.5', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.25', color='black')
ax.set_title("Simple Harmonic Oscillators", 
             fontname="Cambria", fontsize = '14', fontweight = 'bold')
plt.legend(loc = 1, labels = [ 'SHO', 'Underdamped', 'Overdamped'])
plt.show()
```
